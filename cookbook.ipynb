```python
# Assuming 'df' is your DataFrame

# Rename the columns
df = df.rename(columns={
    'Correo electrónico': 'Email',
    'Nombre': 'Name',
    'Where are you from?': 'Location',
    'Name of the recipe': 'Recipe Name',
    'How many people can eat/drink from this?': 'Serving Size',
    'Cooking time': 'Cooking Time',
    'Ingredients and amounts': 'Ingredients',
    'This goes well with...': 'Good Pairings',
    'Feel free to attach pictures of you cooking/eating your dish!': 'Images/ Attachments',
    'Additional comments/notes': 'Additional Comments',
})

# Drop the columns
df = df.drop(columns=['Hora de inicio', 'Hora de finalización', 'What is your name?'])
```

```python
import pandas as pd
from reportlab.lib.pagesizes import letter
from reportlab.platypus import BaseDocTemplate, Frame, PageTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors
from reportlab.lib.units import inch

# Define teal color
teal = colors.HexColor("#007a73")

# Create a BaseDocTemplate
doc = BaseDocTemplate("cookbook.pdf", pagesize=letter)

# Get the styles and customize them
styles = getSampleStyleSheet()
normal_style = styles['BodyText']
normal_style.fontSize = 12

heading_style = styles['Heading1']
heading_style.fontSize = 14

# Function to draw border and set background color for a page
def draw_border_and_bg(canvas, doc):
    canvas.saveState()
    canvas.setStrokeColor(teal)  # border color is teal
    canvas.setFillColor(colors.white)  # background color is white
    canvas.setLineWidth(0.5 * inch)  # Adjust border width
    canvas.rect(0.5 * inch, 0.5 * inch, doc.width + inch, doc.height + inch, fill=True, stroke=True)  # Adjust border margin
    canvas.restoreState()

# Define a frame with padding for the content of a page
frame1 = Frame(doc.leftMargin, doc.bottomMargin, doc.width, doc.height, id='col1',
               leftPadding=0.5*inch, rightPadding=0.5*inch, topPadding=0.5*inch, bottomPadding=0.5*inch)

# Define a page template with the frame and onPage function
template = PageTemplate(id='one_col_template', frames=frame1, onPage=draw_border_and_bg)
doc.addPageTemplates(template)

# Create a story list
story = []

# Iterate over the rows of the DataFrame and add each recipe to the document
for i, row in df.iterrows():
    # Add a Spacer
    story.append(Spacer(1, 12))

    # Add the recipe name as a heading and the other details as normal text
    story.append(Paragraph(row['Recipe Name'], heading_style))
    story.append(Paragraph(f"<strong>Serving Size:</strong> {row['Serving Size']}", normal_style))
    story.append(Paragraph(f"<strong>Cooking Time:</strong> {row['Cooking Time']}", normal_style))
    story.append(Paragraph(f"<strong>Ingredients:</strong> {row['Ingredients']}", normal_style))

    # Break the directions into separate lines with bullet points
    directions = row['Directions'].replace('\n', '<br/>• ')
    story.append(Paragraph(f"<strong>Directions:</strong><br/>• {directions}", normal_style))

    story.append(Paragraph(f"<strong>Good Pairings:</strong> {row['Good Pairings']}", normal_style))
    story.append(Paragraph(f"<strong>Additional Comments:</strong> {row['Additional Comments']}", normal_style))

    # Add a page break after each recipe
    story.append(Spacer(1, 12))
    story.append(PageBreak())

# Build the PDF document with the data added
doc.build(story)
```
